```python
import seaborn as sns
#Import store profile csv
import pandas as pd
import numpy as np
df = pd.read_csv('WELLCOME_APCspend2013_forThinkful.csv', sep=',',header=0)

#Rename column
df.columns = ['ID','Publisher','Journal_Title','Article_Title','Cost']
df.head(2)
```

```python
#Stripping whitespace

df.ID = df.ID.str.strip()
df.Publisher = df.Publisher.str.strip()
df.Journal_Title = df.Journal_Title.str.strip()
df.Article_Title = df.Article_Title.str.strip()
df.Cost = df.Cost.str.strip()
df.Cost = df['Cost'].replace(to_replace='NaN', value=np.nan)
df.Cost = df['Cost'].replace(to_replace='999,999.00', value=np.nan)
df[df.Cost == '999,999.00']
```

```python
#Clean Cost
df['Cost2']=df.Cost.apply(lambda x: ''.join(list(filter(str.isdigit, str(x)))))
df.Cost2=pd.to_numeric(df.Cost2, downcast='float')
df['Cost3'] = np.where(df['Cost2']>0, df.Cost2/100, '')
df.Cost3=pd.to_numeric(df.Cost3, downcast='float')
df.Cost3 = df['Cost3'].replace(to_replace='999,999.00', value=np.nan)
df.head(5)

```

```python
#Capitalize

df.Publisher = df.Publisher.str.upper()
df.Journal_Title = df.Journal_Title.str.upper()
df.Article_Title = df.Article_Title.str.upper()

table=pd.pivot_table(df,index=["Journal_Title"],values=["Article_Title"],aggfunc="count",) 
table_sort = table.sort_values(by='Article_Title', ascending=False)
#Print first 5 lines
print(table_sort.head(5))


```

```python
# Calculate the mean, median, and standard deviation of the open-access cost per article for each journal
table2=pd.pivot_table(df,index=["Journal_Title"],values=["Cost3"],aggfunc="mean",) 
table2=df.sort_values(by='Cost3', ascending=False)
table2.head(5)
```
